## Extracting Article Title, URL, Text and Images in the article

### Using seleniu

```python
#Importing the necessary libraries
import time
import pandas as pd
import nltk
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import StaleElementReferenceException
from selenium.webdriver.common.action_chains import ActionChains
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()))
```

```python
Urls = []
Article_Names = []
Articles_Text = []
Articles_Images = []
```

```python
with open('list-urls.txt', 'r') as file:
    links = [line.strip() for line in file]

# Print the list of links
print(links)
```

```python
for i in range(len(links)):
    url = links[i]
    driver.get(url)
    Urls.append(url)
    print(url)
    time.sleep(10)
    Xpath = "/html/body/form/div[5]/div[1]/div[2]/div[3]/div"
    try:
        Title = driver.find_element(By.XPATH, Xpath).text
    except NoSuchElementException:
        Title = driver.find_element(By.XPATH, "/html/body/form/div[5]/div[2]/div[3]/div").text
    except NoSuchElementException:
        Title = driver.find_element(By.XPATH, "/html/body/form/div[5]/div[2]/div[4]/div/h1").text
    print(Title)
    Article_Names.append(Title)
    time.sleep(5)
    # Assuming driver is already initialized and the page is loaded
    Xpath = "/html/body/form/div[5]/div[1]/div[2]/div[4]"

    # Extract the raw text from the element
    try:
        raw_text = driver.find_element(By.XPATH, Xpath).text
    except NoSuchElementException:
        raw_text = driver.find_element(By.XPATH, "/html/body/form/div[5]/div[2]/div[4]").text
    if '\n' in raw_text:
        formatted_text = "<p>" + raw_text.replace('\n', '</p><p>') + "</p>"
    else:
        formatted_text = f"<p>{raw_text}</p>"
    # Print or use the formatted text as needed
    print(formatted_text)
    Articles_Text.append(formatted_text) 
    
    time.sleep(10)
    try:
        parent_element = driver.find_element(By.XPATH, "/html/body/form/div[5]/div[2]/div[4]")
    except NoSuchElementException:
        parent_element = driver.find_element(By.XPATH, "/html/body/form/div[5]/div[1]/div[2]/div[4]")
    # Find all image elements within the parent element
    img_elements = parent_element.find_elements(By.TAG_NAME, 'img')

    # Extract and store image URLs
    image_urls = []
    for img in img_elements:
        src = img.get_attribute('src')
        if src:
            # Ensure the URL is complete (handling relative URLs)
            if src.startswith('http'):
                image_urls.append(src)
            else:
                full_url = url.join(url, src)
                image_urls.append(full_url)

    # Print image URLs
    print("Image URLs:")
    for url in image_urls:
        print(url)
    formatted_lists = [', '.join(url_list) for url_list in image_urls]
    Articles_Images.append(formatted_lists)
```

```python
data = {
        'Article_Title': Article_Names,
        'Article_URL': Urls,
        'Article_Text': formatted_text
        'Article_Imgs': Article_Images
    }
df = pd.DataFrame(data)
df.to_csv("Mr_Henry_Data1.csv")
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import time

for i in range(len(links)):
    url = links[i]
    driver.get(url)
    Urls.append(url)
    print(url)
    time.sleep(10)

    # Give the page some time to load
    time.sleep(5)

    # Get the page source (HTML content)
    html_content = driver.page_source

    # Parse the HTML using BeautifulSoup
    soup = BeautifulSoup(html_content, 'html.parser')

    # Extract the title (h1 tag)
    title = soup.find('h1').get_text(strip=True)
    print(f"Title: {title}")

    # Extract all paragraph tags and wrap them with <p></p> in formatted text
    paragraphs = soup.find_all('p')
    formatted_text = ''.join([f"<p>{p.get_text(strip=True)}</p>\n" for p in paragraphs])
    print("Formatted Text with <p> tags:")
    print(formatted_text)

    # Extract all image URLs
    image_urls = [img['src'] for img in soup.find_all('img')]
    print("Image URLs:")
    print(image_urls)
```

```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('Mr_Henry_Data1.csv')

# Function to convert a list of links into <img> tags
def convert_links_to_img_tags(links):
    # Check if the cell is NaN or empty, return empty string if it is
    if pd.isna(links):
        return ""
    
    # Split the links if they are comma-separated in the cell
    link_list = links.split(',')  # Assuming the links are comma-separated
    # Convert each link to an <img> tag
    img_tags = [f'<img src="{link.strip()}">' for link in link_list]
    # Join the <img> tags back into a single string
    return ', '.join(img_tags)

# Apply the function to the 4th column and store the result in the 5th column
df['Article_Imgs'] = df.iloc[:, 3].apply(convert_links_to_img_tags)

# Save the modified DataFrame to a new CSV file
df.to_csv('updated_file.csv', index=False)

# Optionally, display the DataFrame to verify the result
print(df)
```
